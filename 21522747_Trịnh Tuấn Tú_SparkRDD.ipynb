```python
import pyspark
from pyspark.sql import SparkSession

spark = SparkSession.builder.getOrCreate()
```

```python
transRDD =  spark.sparkContext.textFile("trans240.csv")
custRDD =  spark.sparkContext.textFile("cust.csv").map(lambda x: x.split(","))
```

```python
transRDD.collect()
```

```python
#9
result9 = transRDD.map(lambda a: a.split(","))\
        .map(lambda a: ((a[2], a[4]), float(a[3])))
result9.collect()
```

```python
result9_1 = result9.reduceByKey(lambda a,b: a+b)\
        .map(lambda a: (a[0][0], (a[0][1], a[1])))\
        .groupByKey()\
        .mapValues(lambda a: sorted(a, key = lambda x: -x[1])[0:3])\
        .sortBy(lambda a: a[0])
result9_1.collect()
```

```python
#10
result10 = transRDD.map(lambda a: a.split(","))
result10.collect()
result10.map(lambda a: (a[4],1))\
            .reduceByKey(lambda a,b: a+b)\
            .sortBy(lambda a: a[1]).collect()
```
